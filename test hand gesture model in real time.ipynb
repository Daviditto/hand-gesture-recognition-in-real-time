```python
import cv2
import mediapipe as mp
from model import HandGestureModel
```

```python
model = HandGestureModel("model_json_1", "model_weights_1.h5")
```

```python
import numpy as np
from PIL import ImageFont, ImageDraw, Image
```

### video test

```python
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 2000)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 2000)

mpHands = mp.solutions.hands
hands = mpHands.Hands(static_image_mode=False,
                      max_num_hands=2,
                      min_detection_confidence=0.5,
                      min_tracking_confidence=0.5)
mpDraw = mp.solutions.drawing_utils

while True:
    ret, img = cap.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
    rx = 1000
    ry = 1000
    black_img = np.zeros(img.shape, dtype="uint8")
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                # print(id,lm)
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                cv2.circle(img, (cx, cy), 3, (255, 0, 255), cv2.FILLED)
                cv2.circle(black_img, (cx, cy), 3, (255, 0, 255), cv2.FILLED)
                if id == 9:
                    rx = int(lm.x * w)
                    ry = int(lm.y * h)
                    cv2.rectangle(img, (rx - 250, ry - 250), (rx + 250, ry + 250), (255, 200, 0), 5)
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
            mpDraw.draw_landmarks(black_img, handLms, mpHands.HAND_CONNECTIONS)

    image = black_img[(ry-250):(ry + 250), (rx - 250):(rx + 250)]
    

    try:
        image = cv2.resize(image, (48, 48))
        label = model.predict_gesture(image[np.newaxis, :, :])
        
        fontpath = "./simsun.ttc"
        font = ImageFont.truetype(fontpath, 64)
        img_pil = Image.fromarray(img)
        draw = ImageDraw.Draw(img_pil)
        draw.text((rx - 250, ry - 240), str(label), font = font, fill = (0,255, 255, 20))
        img = np.array(img_pil)
        #cv2.putText(img, str(label), (rx - 300, ry - 300), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    except:
        print('error')
        
    cv2.imshow("Image", img)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
cv2.waitKey(1)
cap.release()
```
